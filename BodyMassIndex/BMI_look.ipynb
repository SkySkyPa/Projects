```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('bmi_train.csv')
df.head(5)
```

```python
df['Gender'] = pd.factorize(df.Gender)[0]
```

```python
print(df.shape)
print(df.isnull().values.any())
```

```python
sns.jointplot(data=df, x="Height", y="Weight", hue="Index")
```

```python
sns.relplot(data=df, x="Height", y="Weight", hue="Index")
```

```python
sns.scatterplot(data=df, x="Height", y="Weight", hue="Gender")
```

```python
X = df[['Gender', 'Height', 'Weight']]
y = df['Index']
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
predictedDis = regr.predict([[1, 182, 80]])
print(predictedDis)
```
